# Barplot
#### Auf '+' achten. Code ist zwecks Übersichtlichkeit auf verschiedene Zeilen verteilt und mit + verbunden

```r
library(ggplot2)
```

```r
# 1. Grundlagenschicht mit Datensatz und Variable
# Bei Säulendiagramm muss nur Variable für x-Achse angegeben werden.
# Auf y sind die Häufigkeiten
barplot_meer <- ggplot(piraten, aes(x=meer))

# 2. Säulenschicht (hier mit Füllfarbe)
# geom_bar() hat default, dass auf y-Achse die Häufigkeiten abgetragen werden
barplot_meer + 
geom_bar(fill='green') +

# 3. Achsenlabels
labs(x='Meer', y='Häufigkeiten', title='Aufteilung der PiratInnen auf die Meere')
```

## Gruppiertes Säulendiagramm
#### Spezifizierung Geschlecht über Füllfarbe in Grundlagenschicht

```r
barplot_meer <- ggplot(piraten, aes(x=meer, fill=geschlecht))
barplot_meer_gr +
  geom_bar() +
  labs(x='Meer', 
       y='Häufigkeiten', 
       title='Aufteilung der PiratInnen (m/w/d) auf die Meere')
```

#### Spezifizierung Geschlecht in geom_bar, über Füllfarbe der Säulenschicht

```r
barplot_meer <- ggplot(piraten, aes(x=meer)) # ohne Spezifierierung Geschlecht!
barplot_meer +
  
# position='dodge' -> nebeneinander angeordnet statt übereinander
geom_bar(position='dodge',
         aes(fill=geschlecht)) +
labs(x='Meer',
     y='Häufigkeiten',
     title='Aufteilung der PiratInnen (m/w/d) auf die Meere') +

# Optimierung der Labels/Legende mit scale_fill_manual()
scale_fill_manual(values=c('purple','blue','pink'),
                  name='Geschlecht', 
                  breaks = c('d', 'm', 'w'), 
                  labels = c('divers', 'männlich', 'weiblich')) +

# Optimierung der Achsenskalierung
# derzeit gibt es Striche bei 2,5, was halbe Personen bedeuten würde
# scale_y_continuous für y-Achse
scale_y_continuous(breaks = seq(0,20,2)) +

# Ändern der Beschriftungsgröße
theme(axis.text.x = element_text(size = 16),
      axis.text.y = element_text(size = 16),
      legend.text = element_text(size = 16),
      axis.title.x = element_text(size = 20),
      axis.title.y = element_text(size = 20),
      legend.title = element_text(size = 20))
```

#### Code am Stück, ohne Erklärungen...

```r
barplot_meer <- ggplot(piraten, aes(x=meer))

barplot_meer +
geom_bar(position='dodge',
         aes(fill=geschlecht)) +
labs(x='Meer',
     y='Häufigkeiten',
     title='Aufteilung der PiratInnen (m/w/d) auf die Meere') +
scale_fill_manual(values=c('purple','blue','pink'),
                  name='Geschlecht', 
                  breaks = c('d', 'm', 'w'), 
                  labels = c('divers', 'männlich', 'weiblich')) +
scale_y_continuous(breaks = seq(0,20,2)) +
theme(axis.text.x = element_text(size = 16),
      axis.text.y = element_text(size = 16),
      legend.text = element_text(size = 16),
      axis.title.x = element_text(size = 20),
      axis.title.y = element_text(size = 20),
      legend.title = element_text(size = 20))
```

```r
## Beschriftungen sind auch einzeln änderbar:
barplot_meer + 
  geom_bar(fill='black') +
  xlab('Meer') +
  ylab('Häufigkeiten') +
  ggtitle('Aufteilung der PiratInnen auf die Meere')
```
