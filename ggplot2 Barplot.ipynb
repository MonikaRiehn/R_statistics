# Barplot
#### Auf '+' achten. Code ist zwecks Ãœbersichtlichkeit auf verschiedene Zeilen verteilt und mit + verbunden

```r
library(ggplot2)
```

```r
# 1. Grundlagenschicht mit Datensatz und Variable
# Bei SÃ¤ulendiagramm muss nur Variable fÃ¼r x-Achse angegeben werden.
# Auf y sind die HÃ¤ufigkeiten
barplot_meer <- ggplot(piraten, aes(x=meer))

# 2. SÃ¤ulenschicht (hier mit FÃ¼llfarbe)
# geom_bar() hat default, dass auf y-Achse die HÃ¤ufigkeiten abgetragen werden
barplot_meer + 
geom_bar(fill='green') +

# 3. Achsenlabels
labs(x='Meer', y='HÃ¤ufigkeiten', title='Aufteilung der PiratInnen auf die Meere')
```

#### Spezifizierung Geschlecht in geom_bar, Ã¼ber FÃ¼llfarbe der SÃ¤ulenschicht

```r
barplot_meer <- ggplot(piraten, aes(x=meer)) # ohne Spezifierierung Geschlecht!
barplot_meer +
  
# position='dodge' -> nebeneinander angeordnet statt Ã¼bereinander
geom_bar(position='dodge',
         aes(fill=geschlecht)) +
labs(x='Meer',
     y='HÃ¤ufigkeiten',
     title='Aufteilung der PiratInnen (m/w/d) auf die Meere') +

# Optimierung der Labels/Legende mit scale_fill_manual()
scale_fill_manual(values=c('purple','blue','pink'),
                  name='Geschlecht', 
                  breaks = c('d', 'm', 'w'), 
                  labels = c('divers', 'mÃ¤nnlich', 'weiblich')) +

# Optimierung der Achsenskalierung
# derzeit gibt es Striche bei 2,5, was halbe Personen bedeuten wÃ¼rde
# scale_y_continuous fÃ¼r y-Achse
scale_y_continuous(breaks = seq(0,20,2)) +

# Ã„ndern der BeschriftungsgrÃ¶ÃŸe
theme(axis.text.x = element_text(size = 16),
      axis.text.y = element_text(size = 16),
      legend.text = element_text(size = 16),
      axis.title.x = element_text(size = 20),
      axis.title.y = element_text(size = 20),
      legend.title = element_text(size = 20))
```

#### Code am StÃ¼ck, ohne ErklÃ¤rungen...

```r
barplot_meer <- ggplot(piraten, aes(x=meer))

barplot_meer +
geom_bar(position='dodge',
         aes(fill=geschlecht)) +
labs(x='Meer',
     y='HÃ¤ufigkeiten',
     title='Aufteilung der PiratInnen (m/w/d) auf die Meere') +
scale_fill_manual(values=c('purple','blue','pink'),
                  name='Geschlecht', 
                  breaks = c('d', 'm', 'w'), 
                  labels = c('divers', 'mÃ¤nnlich', 'weiblich')) +
scale_y_continuous(breaks = seq(0,20,2)) +
theme(axis.text.x = element_text(size = 16),
      axis.text.y = element_text(size = 16),
      legend.text = element_text(size = 16),
      axis.title.x = element_text(size = 20),
      axis.title.y = element_text(size = 20),
      legend.title = element_text(size = 20))
```
