## Gruppierter Boxplot, mit Subset ohne .na
Boxplot für die Gefährlichkeit der PiratInnen, unterschieden nach Schwimmern und Nichtschwimmern. Auf x-Achse wird nach Schwimmern und Nichtschwimmern getrennt, auf y-Achse wird Gefährlichkeit abgetragen.

```r
# 1. Grundlagenschicht
# hier aus einen Subset ohne Piraten, bei denen Angabe zu Schwimmfähigkeit fehlt
boxplot_gefahr <- ggplot(piraten[!is.na(piraten$schwimmen), ], aes(x=schwimmen, y=gefahr))

# 2. Boxschicht
boxplot_gefahr +
stat_boxplot(geom='errorbar', width=0.5) +
geom_boxplot() +

# 3. Achsenbeschriftung
labs(x='Schwimmfähigkeit',
    y='Gefährlichkeit',
    title='Boxplot für die Gefährlichkeit der PiratInnen, getrennt nach Schwimmer und NichtschwimmerInnen')
```

#### Interpretation des Boxplots
